```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

df = pd.read_csv("/content/crime_safety_dataset.csv")
df.info()
crime_counts = df['crime_type'].value_counts()
```

```python
print("Top 5 Most Common Crimes:")
print(crime_counts.head(10))
```

```python
top_crimes = df['crime_type'].value_counts().nlargest(5).index
```

```python
sns.countplot(data=df[df['crime_type'].isin(top_crimes)],x='crime_type', hue='victim_gender')
plt.xticks(rotation=45,ha='right')
plt.show()
```

```python
crimes = sns.catplot(data=df, x='crime_type', col='victim_gender', kind='count', palette='viridis')
crimes.set_xticklabels(rotation=90)
plt.tight_layout()
plt.show()
```

```python
bins = [0, 10, 20, 30, 40, 50, 60, 70, 100]
labels = ['0-10', '11-20', '21-30', '31-40', '41-50', '51-60', '61-70', '71+']
df['age_group'] = pd.cut(df['victim_age'], bins=bins, labels=labels)
```

```python
top_crimes = df['crime_type'].value_counts().head(5).index
filtered_df = df[df['crime_type'].isin(top_crimes)]
```

```python
sns.countplot(data=filtered_df, x='age_group', hue='crime_type')
plt.title("Top 5 Crimes by Victim Age Group")
plt.xlabel("Victim Age Group")
plt.ylabel("Number of Crimes")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
